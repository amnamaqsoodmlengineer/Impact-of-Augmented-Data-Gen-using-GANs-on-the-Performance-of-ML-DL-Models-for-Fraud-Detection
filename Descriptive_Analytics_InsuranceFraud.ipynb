Skewness and Kurtosis

Skewness is a measure of the asymmetry of a data distribution, while kurtosis is a measure of the peakedness of a data distribution. Both measures can help identify underlying patterns in data that can indicate potential outliers or errors in data collection.

What Is Skewness?
Skewness is a statistical measure that assesses the asymmetry of a probability distribution. It quantifies the extent to which the data is skewed or shifted to one side.

Positive skewness indicates a longer tail on the right side of the distribution, while negative skewness indicates a longer tail on the left side. Skewness helps in understanding the shape and outliers in a dataset.

---
What Is Kurtosis?
Kurtosis is a statistical measure that quantifies the shape of a probability distribution. It provides information about the tails and peakedness of the distribution compared to a normal distribution.

Positive kurtosis indicates heavier tails and a more peaked distribution, while negative kurtosis suggests lighter tails and a flatter distribution. Kurtosis helps in analyzing the characteristics and outliers of a dataset.

The measure of Kurtosis refers to the tailedness of a distribution. Tailedness refers to how often the outliers occur.

Display a Density Plot

KDE Density Plot: A KDE density plot is a graphical representation of data that displays the probability density of a given data set. It is a smooth curve that shows the estimated probability density of each data point.

```python
# df.policy_annual_premium.plot.kde()
import seaborn as sns
import matplotlib.pyplot as plt

# List of features to plot
features_to_plot = ['policy_annual_premium', 'injury_claim', 'property_claim']

# Draw density plots for each feature
for feature in features_to_plot:
    sns.kdeplot(df[feature], shade=True, label=feature)

# Add labels and title
plt.xlabel('Feature Values')  # X-axis label
plt.ylabel('Density')          # Y-axis label
plt.title('Density Plot of Features')  # Title

# Add legend
plt.legend()

# Display the plot
plt.show()
```

isplay an Area Plot

An area plot is a two-dimensional graph that displays quantitative data through a filled polygon, which is created by connecting a sequence of points. It is used to visualise the changes in data over a certain period of time, track cumulative totals, and compare the relationship between two or more variables.

```python
 #df.plot.area()
import matplotlib.pyplot as plt

# Selecting the features for the area plot
features_to_plot = ['policy_annual_premium', 'injury_claim', 'property_claim']

# Plot the area plot for the selected features
df[features_to_plot].plot.area()

# Add labels and title
plt.xlabel('Index')  # X-axis label
plt.ylabel('Feature Values')  # Y-axis label
plt.title('Area Plot of Top 3 Features')  # Title

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Set up the figure and axis grid
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Flatten the axis grid for easy iteration
axs = axs.flatten()

# Plot histograms for each feature
for i, feature in enumerate(['months_as_customer', 'age', 'policy_annual_premium',
                             'incident_days', 'incident_hour_of_the_day', 'total_claim_amount']):
    axs[i].hist(df[feature], bins=20, color='skyblue', edgecolor='black')
    axs[i].set_title(feature.capitalize())
    axs[i].set_xlabel('Value')
    axs[i].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
